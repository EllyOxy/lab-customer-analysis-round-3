```python
#For the numerical variables, check the multicollinearity between the features. #
#Please note that we will use the column total_claim_amount later as the target variable.
```

```python
correlations_matrix = numerical.corr() # method="spearman"
correlations_matrix

```

```python
correlations_matrix["total_claim_amount"].sort_values(ascending=False)
```

```python
correlations_matrix[(correlations_matrix['total_claim_amount'] > .1) | (correlations_matrix['total_claim_amount'] < -.1)]['total_claim_amount'].sort_values(ascending=False)
```

```python
mask = np.zeros_like(correlations_matrix)
mask[np.triu_indices_from(mask)] = True 

fig, ax = plt.subplots(figsize=(6, 4))
ax = sns.heatmap(correlations_matrix, mask=mask, annot=True)
plt.show()
```
